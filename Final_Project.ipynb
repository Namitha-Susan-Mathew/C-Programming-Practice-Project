<h3>Checking For And Removing Duplicate Values</h3>
<p>Duplicate entries need to be removed to ensure data integrity and avoid redundancy in analysis. The dataset was checked for duplicate records using the duplicated() function. No significant duplicates were found; however, duplicates were removed as a precautionary step to maintain data integrity.</p>

```python
# Check duplicates
df.duplicated().sum()

# Remove duplicates
df = df.drop_duplicates()
```

<h3>Handling Missing Values</h3>
<p>Missing values in the columns need to be handled. Teacher_Quality has 78 missing values, Parental_Education_Level has 90 missing values, and Distance_from_Home has 67 missing values. All other columns do not contain missing data.</p>
<p>Missing values were replaced with ‘Unknown’ to maintain data integrity and explicitly represent unavailable information. A verification was also done to ensure that no missing values are present in the dataset.</p>

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing values with 'Unknown'
df['Teacher_Quality'].fillna('Unknown', inplace=True)
df['Parental_Education_Level'].fillna('Unknown', inplace=True)
df['Distance_from_Home'].fillna('Unknown', inplace=True)
```

```python
# Verify
df.isnull().sum()
```

<h3>Aggregating Data</h3>
<p>Aggregation technique such as groupby() were used to compute summary statistics. This helps in understanding patterns and comparing performance across different categories.</p>
<p>This was used to calculate the average score based on Gender and Motivation_Level.</p>

```python
# Average Exam Score by Gender
avg_score_gender=df.groupby('Gender')['Exam_Score'].mean()
avg_score_gender
```

<p>Aggregation was performed using the groupby() function to calculate the average exam score for each motivation level. This helps in summarizing the data and understanding how student performance varies across different levels of motivation.</p>

```python
# Average Score by Motivation Level
avg_score_motivation = df.groupby('Motivation_Level')['Exam_Score'].mean()
avg_score_motivation
```

<p>Aggregation was performed using the groupby() function to calculate the average exam score corresponding to the education level of the student's parents.This helps to summarize the data and understand how the parent's education level affect student performance.</p>

```python
# Average score by Parent Involvement
avg_score_parent = df.groupby('Parental_Involvement')['Exam_Score'].mean()
avg_score_parent
```

<p>Aggregation was performed using the groupby() function to calculate the average exam score with respect to the students'access to the internet. This helps to summarize the data and understand how the access to internet and resources affect student performance.



```python
#Average Score by Internet Access
avg_score_internet= df.groupby('Internet_Access')['Exam_Score'].mean()
avg_score_internet
```

<p>Aggregation was performed using the groupby() function to calculate the average exam score with respect to students’ family income levels. This helps to summarize the data and understand how economic background influences student performance by comparing average scores across different income groups.</p>

```python
#Average Score by Family Income
avg_score_income = df.groupby('Family_Income')['Exam_Score'].mean()
avg_score_income
```

<h2>Feature Engineering</h2>
<p>New meaningful columns need to be derived from the existing columns in the dataset for better analysis, improved visualizations and clearer insights.These engineered features enable deeper analysis and help uncover meaningful patterns in student performance.<p>
<p><b>Performance Level</b> - A new column called "Performance_Level" was created based on the exam score to categorize students into three groups: High, Medium, and Low performers. This helps simplify analysis and enables easier comparison across different performance groups.

```python
df['Performance_Level'] = df['Exam_Score'].apply(
    lambda x: 'High' if x >= 75 else ('Medium' if x >= 50 else 'Low')
)
df['Performance_Level']
```

<p><b>Study_Efficiency</b> - A new feature called "Study_Efficiency" was created by dividing the exam score by the number of hours studied. This metric helps evaluate how effectively students utilize their study time.

```python
df['Study_Efficiency'] = df['Exam_Score'] / df['Hours_Studied']
df['Study_Efficiency']
```

<h3>Distribution Of Exam Scores</h3>
<p>This graph helps us understand how students are performing overall. It shows whether most students are scoring high, low, or somewhere in the middle.</p>

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], bins=20)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students")
plt.show()
```

<h5><b>Insights</b></h5>
<ul>
<li>The majority of students scored between 65 and 72, indicating that the average score likely falls within this range. The peak (mode) appears to be around 68–69. This indicates that majority of the students are average performers.</li>
<li>Scores below 60 and above 80 are minimal, also, most scores are concentrated between 60 and 75, suggesting that the exam was neither too easy nor too difficult for most students.</li>
<li>There are a small number of students who scored above 90, which could indicate exceptional performance.</li>
</ul>

<h3>Relationship between Study Hours and Exam Score</h3>
<p>This graph is used to analyze relationship between study time and exam performance. This graph helps us see if studying more actually leads to better marks.</p>

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

<p><b>Insights</b></p>
<ul>
<li>The scatter plot shows a positive relationship between study hours and exam scores, meaning that in general, students who study more tend to score higher, but the relationship is not very strong.</li>
<li>Beyond a certain saturation point (20-25 hours), additional study time may not significantly improve performance.</li>
<li>Study time helps, but it is not the only factor. Some students have scored high marks despite studying for a few hours, whereas some others have scored low marks even after studying for more hours. This shows that study hours alone are not a perfect predictor of exam performance.</li>
